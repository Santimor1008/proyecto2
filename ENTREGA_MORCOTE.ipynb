### Importación de datos



```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_t1 = sum(tienda['Precio'])
ingreso_total_t2 = sum(tienda2['Precio'])
ingreso_total_t3 = sum(tienda3['Precio'])
ingreso_total_t4 = sum(tienda4['Precio'])

print(f'El ingreso total de la tienda 1 es: {ingreso_total_t1}')
print(f'El ingreso total de la tienda 2 es: {ingreso_total_t2}')
print(f'El ingreso total de la tienda 3 es: {ingreso_total_t3}')
print(f'El ingreso total de la tienda 4 es: {ingreso_total_t4}')
#print(tienda['Precio'])
```

# 2. Ventas por categoría

```python
def agregar_a_lista(numero_tienda):
  lista = []
  categorias = []
  for categoria in numero_tienda['Categoría del Producto']:
    categorias.append(categoria)
    if categoria not in lista:
      lista.append(categoria)
  for i in lista:
    categorias_contador = categorias.count(i)
    print(i, categorias_contador)

print("-----TIENDA 1----")
agregar_a_lista(tienda)
print("----TIENDA 2-----")
agregar_a_lista(tienda2)
print("-----TIENDA 3-----")
agregar_a_lista(tienda3)
print("------TIENDA 4-----")
agregar_a_lista(tienda4)

```

# 3. Calificación promedio de la tienda


```python
def calcular_promedio_de_calificacion(numero_tienda, tienda_name):
  suma_total = sum(numero_tienda['Calificación'])
  cantidad_de_calificaciones = len(numero_tienda['Calificación'])
  #print(cantidad_de_calificaciones, suma_total)
  return(f'EL PROMEDIO DE CALIFICACION DE LA {tienda_name} ES {suma_total/cantidad_de_calificaciones}')
calificaciones_tienda_1 = calcular_promedio_de_calificacion(tienda, 'Tienda 1')
calificaciones_tienda_2 = calcular_promedio_de_calificacion(tienda2, 'Tienda 2')
calificaciones_tienda_3 = calcular_promedio_de_calificacion(tienda3, 'Tienda 3')
calificaciones_tienda_4 = calcular_promedio_de_calificacion(tienda4, 'Tienda 4')

print(calificaciones_tienda_1)
print(calificaciones_tienda_2)
print(calificaciones_tienda_3)
print(calificaciones_tienda_4)
```

# 4. Productos más y menos vendidos

```python
def listar_cantidad_de_produtos(numero_tienda):
  diccionario_contar = {}
  productos = []
  for producto in numero_tienda['Producto']:
    productos.append(producto)
    diccionario_contar[producto] = productos.count(producto)
    resultado_minimo = [nombre for nombre, cantidad in diccionario_contar.items() if cantidad == min(diccionario_contar.values()) ]
    resultado_maximo = [nombre for nombre, cantidad in diccionario_contar.items() if cantidad == max(diccionario_contar.values()) ]
    cantidad = {}

    for i in resultado_minimo:
      cantidad[i] = diccionario_contar[i]
    for i in resultado_maximo:
      cantidad[i] = diccionario_contar[i]
  maximos = f"Los productos mas vendidos son: {resultado_maximo}"
  minimos = f"Los productos menos vendidos son: {resultado_minimo}"
  claves = []
  valores = []
  for nombre,cantidad_de_productos in cantidad.items():
    claves.append(nombre)
    valores.append(cantidad_de_productos)
  plt.figure(figsize=(15, 6))
  plt.bar(x=claves,height=valores)
  plt.show
  return(maximos,minimos)

cantidad_de_productos = listar_cantidad_de_produtos(tienda)
print("TIENDA 1",cantidad_de_productos)
cantidad_de_productos_2 = listar_cantidad_de_produtos(tienda2)
print("TIENDA 2",cantidad_de_productos_2)
cantidad_de_productos_3 = listar_cantidad_de_produtos(tienda3)
print("TIENDA 3",cantidad_de_productos_3)
cantidad_de_productos_4 = listar_cantidad_de_produtos(tienda4)
print("TIENDA 4",cantidad_de_productos_4)

```

# 5. Envío promedio por tienda

```python
def calcular_promedio_de_envio(numero_tienda, tienda_name):
  suma_total = sum(numero_tienda['Costo de envío'])
  cantidad_de_envios = len(numero_tienda['Costo de envío'])
  #print(cantidad_de_envios, suma_total)
  return(f'EL PROMEDIO DE COSTO POR ENVIOS DE LA {tienda_name} ES {suma_total/cantidad_de_envios}')
envios_tienda_1 = calcular_promedio_de_envio(tienda, 'Tienda 1')
envios_tienda_2 = calcular_promedio_de_envio(tienda2, 'Tienda 2')
envios_tienda_3 = calcular_promedio_de_envio(tienda3, 'Tienda 3')
envios_tienda_4 = calcular_promedio_de_envio(tienda4, 'Tienda 4')

print(envios_tienda_1)
print(envios_tienda_2)
print(envios_tienda_3)
print(envios_tienda_4)
```

# 6. GRAFICOS UTILES





```python
# DISTRIBUCIÓN POR CATEGORIAS
def agregar_a_lista(numero_tienda,nombre_tienda):
  lista = []
  categorias = []
  categorias_contador = []
  for categoria in numero_tienda['Categoría del Producto']:
    categorias.append(categoria)
    if categoria not in lista:
      lista.append(categoria)
  #print(lista,categorias)
  for i in lista:
    categorias_contador.append(categorias.count(i))
  plt.pie(categorias_contador, labels = lista, autopct= "% 0.1f%%")
  plt.title(nombre_tienda)
  plt.show()


  #return(lista, categorias_contador)

#print("-----TIENDA 1----")
agregar_a_lista(tienda,"TIENDA 1")
agregar_a_lista(tienda2,"TIENDA 2")
agregar_a_lista(tienda3,"TIENDA 3")
agregar_a_lista(tienda4,"TIENDA 4")
```

**METODO DE PAGO MAS USADO**




```python
def contar_metodos_de_pago(numero_tienda,nombre_tienda):
  lista = []
  metodo_de_pago = []
  contador = []
  for metodo in numero_tienda['Método de pago']:
    metodo_de_pago.append(metodo)
    if metodo not in lista:
      lista.append(metodo)
  #print(lista,categorias)
  for i in lista:
    contador.append(metodo_de_pago.count(i))
  plt.barh( y= lista, width=contador)
  plt.title(nombre_tienda)
  plt.show()

contar_metodos_de_pago(tienda,"TIENDA 1")
contar_metodos_de_pago(tienda2,"TIENDA 2")
contar_metodos_de_pago(tienda3,"TIENDA 3")
contar_metodos_de_pago(tienda4,"TIENDA 4")
```
